```python
%pylab inline
%load_ext cythonmagic

import mahotas as mh
import numpy as np
import os
import json
import cv2
from time import time
from math import radians
from skimage import io, color, img_as_ubyte
from skimage.transform import rescale
from skimage.morphology import square
from skimage.filter import gabor_filter, gaussian_filter
from sklearn.decomposition import PCA, RandomizedPCA
from sklearn.cluster import DBSCAN
from scipy.cluster.vq import kmeans2
from sklearn.svm import SVC
from sklearn.feature_selection import RFE

from helpers import display as dh
from helpers import features as fh

BASE_PATH = os.getcwd()
print("Current base path: {0}".format(BASE_PATH))
DATA_PATH = BASE_PATH + '/Daten/2D/Talus_dorsal_cut_2/'
TO_PATH = BASE_PATH + '/Daten/2D/Talus_dorsal_outlined/'
TEST_FILES = json.loads(open(BASE_PATH + '/Daten/2D/testfiles.json').read())

ALPHA = 0.25
BETA = 0.5

def get_gabor_kernels():
    kernels = []
    angles = [ radians(i) for i in [ 0, 30, 60, 90, 120, 150 ] ]
    #angles = [ radians(i) for i in [ 0, 45, 90, 135 ] ]
    #frequencies = [ sqrt(2) * i for i in [ 1, 2, 4, 8, 16, 32, 128, 256 ] ] # Jahn
    base = [ 1, 2, 4, 8, 16, 32, 64, 128 ]
    frequencies = [ 0.25 - (2 ** (i-0.5)) / 1024 for i in base ] + [ 0.25 + (2 ** (i-0.5)) / 1024 for i in base ]# Zhang
    #frequencies = [ 0.1, 0.25, 0.4, 1]
    
    for angle in angles:
        for frequency in frequencies:
            kernels.append((frequency, angle))
            
    return kernels

def apply_gabor_kernels(image, kernels):
    number_of_kernels = len(kernels)
    third_dimension = 1 if len(image.shape) != 3 else image.shape[2]
    features = np.zeros((image.shape[0], image.shape[1], number_of_kernels * third_dimension))
    for i, kernel in enumerate(kernels):
        print('Processing Kernel {0} of {1}'.format(i+1, len(kernels)))
        for j in range(third_dimension):
            feature_index = i*third_dimension + j
            image_dim = image if third_dimension == 1 else image[:,:,j]
            features[:,:,feature_index] = gabor_filter(image_dim, kernel[0], kernel[1])[0]
            features[:,:,feature_index] = fh.normalize(features[:,:,feature_index])
    return features

def apply_nonlinearity(features):
    return np.tanh(np.multiply(features, ALPHA))

def add_locality_info(features):
    locality = zeros((features.shape[0], features.shape[1], 2))
    for i in range(features.shape[0]):
        for j in range(features.shape[1]):
            locality[i,j,0] = i / features.shape[0]
            locality[i,j,1] = j / features.shape[1]
    return np.concatenate((features, locality), 2)

def apply_blur(features, kernels):
    for i, kernel in enumerate(kernels):
        window_size = 25
        sigma = (1024 / kernel[0]) * BETA
        features[:,:,i] = cv2.GaussianBlur(features[:,:,i], (51, 51), sigma)
        features[:,:,i] = fh.normalize(features[:,:,i])
    return features

def process_features_for_single_image(image):
    ascii_filename = file['filename'].encode('ascii', 'replace')

    start_time = time()
    kernels = get_gabor_kernels()
    print('{0} get_gabor_kernels: {1}s'.format(ascii_filename, time() - start_time))

    start_time = time()
    features = apply_gabor_kernels(image, kernels)
    print('{0} apply_gabor_kernels: {1}s'.format(ascii_filename, time() - start_time))
    
    start_time = time()
    features = apply_nonlinearity(features)
    print('{0} apply_nonlinearity: {1}s'.format(ascii_filename, time() - start_time))
    
    start_time = time()
    #features = apply_blur(features, kernels)
    print('{0} apply_blur: {1}s'.format(ascii_filename, time() - start_time))
    
    start_time = time()
    features = add_locality_info(features)
    print('{0} add_locality_info: {1}s'.format(ascii_filename, time() - start_time))
    
    return features.reshape((features.shape[0] * features.shape[1], features.shape[2]))

def reduce_feature_complexity(features):
    #pca = PCA(n_components=3)
    pca = RandomizedPCA(n_components=6)
    reduced = pca.fit_transform(features)
    
    return reduced
  
def do_classification(features):
    clf = SVC(kernel='linear')
    
    flattened = features.reshape((features.shape[0] * features.shape[1], features.shape[2]))
    
    background_indices_rows = np.array(file['background'])[:,0]
    background_indices_columns = np.array(file['background'])[:,1]
    background = features[background_indices_rows, background_indices_columns]
    background_class = np.zeros(background.shape[0])
    
    foreground_indices_rows = np.array(file['foreground'])[:,0]
    foreground_indices_columns = np.array(file['foreground'])[:,1]
    foreground = features[foreground_indices_rows, foreground_indices_columns]
    foreground_class = np.ones(foreground.shape[0])
    
    features_for_known_classes = np.concatenate((background, foreground))
    known_classes = np.concatenate((background_class, foreground_class))
    
    start_time = time()
    clf.fit(features_for_known_classes, known_classes)
    print('fit: {0}s'.format(time() - start_time))
    
    return clf.predict(flattened).reshape((features.shape[0], features.shape[1]))

def do_clustering(features):
    flattened = features.reshape((features.shape[0] * features.shape[1], features.shape[2]))
    #db = DBSCAN(eps=0.3, min_samples=50).fit(flattened)
    #labels = db.labels_
    centroids, labels = kmeans2(flattened, 4)
    return labels.reshape((features.shape[0], features.shape[1]))
```
