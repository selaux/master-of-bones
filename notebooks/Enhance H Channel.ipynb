```python
%pylab nbagg

import numpy as np
import os
import json
from functools import partial
from skimage import color, io
from skimage.transform import rescale
from scipy.ndimage.measurements import histogram as histo

from helpers import display as dh

BASE_PATH = os.getcwd()
print("Current base path: {0}".format(BASE_PATH))
DATA_PATH = BASE_PATH + '/Daten/2D/Talus_dorsal_cut_2/'
TO_PATH = BASE_PATH + '/Daten/2D/Talus_dorsal_outlined/'
TEST_FILES = json.loads(open(BASE_PATH + '/Daten/2D/testfiles.json').read())

def is_closer(vbone, voutside, vinquestion):
    dist_bone = np.linalg.norm(vbone - vinquestion)
    dist_outside = np.linalg.norm(voutside - vinquestion)
    return 1 if dist_bone < dist_outside else 0

def calculate_single_image(rgb_image):
    hue_channel = color.rgb2hsv(rgb_image)[:,:,0]
    num_pixels = hue_channel.shape[0] * hue_channel.shape[1]
    
    fig, ax = plt.subplots(1, 2)
    ax[0].imshow(rgb_image)
    ax[0].set_title("RGB")
    ax[1].imshow(hue_channel)
    ax[1].set_title("Clusters")
    histogram, bin_borders = np.histogram(hue_channel, bins=64)
    ibone = None
    ioutside = None
    
    for i in range(len(histogram)):
        ibefore = i-1
        iafter = i+1
        if ibefore > 0 and iafter < len(histogram):
            v = histogram[i]
            vbefore = histogram[ibefore]
            vafter = histogram[iafter]
            if vbefore < v and vafter < v and v > num_pixels / 50:
                if ibone == None:
                    ibone = i
                elif ioutside == None:
                    ioutside = i
                
    vbone = bin_borders[ibone-1] + (bin_borders[ibone] - bin_borders[ibone-1]) / 2
    voutside = bin_borders[ioutside-1] + (bin_borders[ioutside] - bin_borders[ioutside-1]) / 2
    
    flattened = hue_channel.reshape((hue_channel.shape[0] * hue_channel.shape[1]))
    pixel_labels = np.array(map(partial(is_closer, vbone, voutside), flattened))
    
    return pixel_labels.reshape(hue_channel.shape[0], hue_channel.shape[1])
```

```python
file = TEST_FILES[7]

rgb_image = io.imread(DATA_PATH + file['filename'])
rgb_image = rescale(rgb_image, .25)
filtered = calculate_single_image(rgb_image)
dh.clusters(plt, rgb_image, filtered)
```
