```python
%pylab inline

import numpy as np
import os
import cv2
import json
from time import time
from math import radians
from skimage import io, color, img_as_float
from skimage.transform import rescale
from skimage.util import pad
from IPython.parallel import Client
from sklearn.decomposition import PCA, RandomizedPCA
from scipy.cluster.vq import kmeans2
from sklearn.svm import SVC

import helpers.features as fh
import helpers.display as dh

BASE_PATH = os.getcwd()
print("Current base path: {0}".format(BASE_PATH))
DATA_PATH = BASE_PATH + '/Daten/2D/Talus_dorsal_cut_2/'
TRAINING_PATH = BASE_PATH + '/Daten/2D/Training/'
TO_PATH = BASE_PATH + '/Daten/2D/Talus_dorsal_outlined/'
TEST_FILES = json.loads(open(BASE_PATH + '/Daten/2D/testfiles.json').read())

WINDOW_SIZE = 10
PERCENTAGE_OF_TRAINING = .8

def extract_semi_locality_for_each_pixel(image):
    step_size = WINDOW_SIZE / 2
    image_with_padded_borders = pad(image, ((step_size, step_size), (step_size, step_size), (0, 0)), mode='symmetric')
    neighborhoods = np.zeros((image.shape[0] * image.shape[1], WINDOW_SIZE, WINDOW_SIZE, 3))
    
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            neighborhoods[y*image.shape[1] + x] = image_with_padded_borders[y:y+WINDOW_SIZE, x:x+WINDOW_SIZE]
            
    return neighborhoods

def map_feature_for_pixel(neighborhood):
    import numpy as np
    from math import exp
    from scipy.ndimage.filters import convolve
    
    SIGMA = .1
    
    gradient_filter_y = [ [0, 0, 0], [-1, 0, 1], [0, 0, 0] ]
    gradient_images_y = np.zeros_like(neighborhood)
    gradient_images_y[:,:,0] = convolve(neighborhood[:,:,0], gradient_filter_y)
    gradient_images_y[:,:,1] = convolve(neighborhood[:,:,1], gradient_filter_y)
    gradient_images_y[:,:,2] = convolve(neighborhood[:,:,2], gradient_filter_y)
    
    gradient_filter_x = [ [0, -1, 0], [0, 0, 0], [0, 1, 0] ]
    gradient_images_x = np.zeros_like(neighborhood)
    gradient_images_x[:,:,0] = convolve(neighborhood[:,:,0], gradient_filter_x)
    gradient_images_x[:,:,1] = convolve(neighborhood[:,:,1], gradient_filter_x)
    gradient_images_x[:,:,2] = convolve(neighborhood[:,:,2], gradient_filter_x)
    
    result = np.zeros((neighborhood.shape[0], neighborhood.shape[1]))
    for y in range(neighborhood.shape[0]):
        for x in range(neighborhood.shape[1]):
            mat = np.array([
                [ 1 + np.sum(np.power(gradient_images_x[y,x,:], 2)), np.sum(np.multiply(gradient_images_x[y,x,:], gradient_images_y[y,x,:])) ],
                [ np.sum(np.multiply(gradient_images_x[y,x,:], gradient_images_y[y,x,:])), 1 + np.sum(np.power(gradient_images_x[y,x,:], 2)) ]
            ])
            result[y,x] = exp(-np.linalg.det(mat) / SIGMA)
    return result.flatten()
    
def process_features_for_single_image(image):
    neighborhoods = extract_semi_locality_for_each_pixel(image)
    
    #rc = Client(profile='home')
    #lview = rc.load_balanced_view()
    #lview.block = True
    features = np.array(map(map_feature_for_pixel, neighborhoods))
    #features = np.array(lview.map(map_feature_for_pixel, neighborhoods))
    
    features = fh.normalize_each(features)
    
    return features
        
def reduce_feature_complexity(features):
    #pca = PCA(n_components=3)
    pca = RandomizedPCA(n_components=6)
    reduced = pca.fit_transform(features)
    return reduced

def do_clustering(image, features):
    #db = DBSCAN(eps=0.5, min_samples=10).fit(flattened)
    #labels = db.labels_
    centroids, labels = kmeans2(features, 4)
    return labels.reshape((image.shape[0], image.shape[1]))
    
def do_classification(image, training_mask, features):
    clf = SVC()
    
    training_length_for_each_class = int(PERCENTAGE_OF_TRAINING * features.shape[0] / 2)
    background_classes = zeros((training_length_for_each_class))
    foreground_classes = ones((training_length_for_each_class))
    
    background_features = features[(training_mask == 0).flatten()]
    foreground_features = features[(training_mask == 1).flatten()]
    bg_indices = np.random.choice(range(background_features.shape[0]), training_length_for_each_class)
    fg_indices = np.random.choice(range(foreground_features.shape[0]), training_length_for_each_class)
    
    print(bg_indices)
    print(fg_indices)
    
    features_for_known_classes = np.concatenate((
        background_features[bg_indices, :],
        foreground_features[fg_indices, :]
    ))
    known_classes = np.concatenate((background_classes, foreground_classes))
    
    start_time = time()
    clf.fit(features_for_known_classes, known_classes)
    print('fit: {0}s'.format(time() - start_time))
    
    return clf.predict(features).reshape((image.shape[0], image.shape[1]))
```

```python
file = TEST_FILES[0]
SCALE = .1
rgb_image = io.imread(DATA_PATH + file['filename'])
rgb_image = rescale(rgb_image, SCALE)
training_mask = io.imread(TRAINING_PATH + file['filename'])
training_mask = color.rgb2gray(rescale(training_mask, SCALE))
training_mask[training_mask > .8] = 1
training_mask[training_mask <= .8] = 0

features = process_features_for_single_image(rgb_image)
```
